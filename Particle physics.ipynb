```python
from vpython import *
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# Create a NEW canvas each time you run
canvas()  # Add this line!

E=np.array([1,2,3])
m=1
q=1
t=np.linspace(0,10,1000)
def F(state,t):
    x,y,z,vx,vy,vz=state
    ax=E.T[0]*(q/m)
    ay=E.T[1]*(q/m)
    az=E.T[2]*(q/m)
    
    return [vx,vy,vz,ax,ay,az]

state0=[0,0,0,0,0,0]
sol=odeint(F,state0,t)

# Create figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Your plotting code here

ax.plot(sol.T[0],sol.T[1],sol.T[2],color="orange")
    
    
'''r=[x,y,z]
r1=np.array(r)

ball=sphere(pos=vector(r1[0],r1[1],r1[2]),radius=1,color=color.orange)'''
```
